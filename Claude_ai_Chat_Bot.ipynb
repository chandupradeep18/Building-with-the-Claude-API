```python
from anthropic import Anthropic
from google.colab import userdata
import csv
import os
from datetime import datetime
```

```python
client=Anthropic(api_key=userdata.get('Claude'))
model='claude-haiku-4-5'
max_token=500
```

```python
chat_history=list()
def user_message(message):
  user={"role":"user","content":message}
  details={"Date & time":datetime.now(),"role":"user","content":message}
  chat_history.append(user)
  Collect_response_of_chat(details)

def assistant_message(message):
  bot={"role":"assistant","content":message}
  details={"Date & time":datetime.now(),"role":"assistant","content":message}
  chat_history.append(bot)
  Collect_response_of_chat(details)

def chat(message,system=None):
  param={
      "model":model,
      "max_tokens":max_token,
      "messages":message
  }
  if system:
    param["system"]=system
  return client.messages.create(**param)
def Collect_Response_detail(details):
  file_exist=os.path.exists('response_details.csv')
  with open('response_details.csv','a', newline="") as csv_file:
    writer=csv.DictWriter(csv_file,fieldnames=details.keys())
    if not file_exist:
        writer.writeheader()
    writer.writerow(details)
def Collect_response_of_chat(details):
  file_exist=os.path.exists('chat_history.csv')
  with open('chat_history.csv','a',newline="") as csv_file:
    writer=csv.DictWriter(csv_file,fieldnames=details.keys())
    if not file_exist:
      writer.writeheader()
    writer.writerow(details)
```

```python
chat_history=list()
bot_role=input("Enter a Bot Role : ")
while True:
  user_input=input("Ask Ai any Thing : ")
  if not user_input:
    break
  system=None
  if bot_role:
    system=bot_role
  user_message(user_input)
  response=chat(chat_history,system)
  assistant_message(response.content[0].text)
  print(response.content[0].text)
  details={'Time':datetime.now(),'id':response.id,'role':response.role,'stop_reason':response.stop_reason,'input_tokens':response.usage.input_tokens,'output_tokens':response.usage.output_tokens}
  Collect_Response_detail(details)
```
